```python
import pandas as pd
import glob
import os
import numpy as np
#先删除原来的图片
photo=glob.glob("*.png")
for _ in photo:
    os.remove(_)
os.chdir("E:/Bi_ye/1型层间断裂韧性原始数据/1型层间断裂韧性原始数据/")
file_name=glob.glob("*-2.csv")
for file in file_name:
    data=pd.read_csv(file,encoding="gbk",skiprows=3,header=None,dtype=float)
    del data[4],data[5]
    data=data[data[2]>0]
    if max(data[3])>40:
        photo_name=os.path.splitext(file)[0]
        plt.ylabel("Load(N)")
        xlim=np.arange(0,40,5)
        ylim=np.arange(0,90,5)
        plt.xticks(xlim)
        plt.yticks(ylim)
        plt.xlabel("Displacement(mm)")
        plt.title("Load-Displacement")
        plt.plot(data[2],data[3],"r",label=photo_name)
        plt.legend()
        plt.grid()
        plt.savefig(photo_name+'.png')
        plt.close()
    
    
```
