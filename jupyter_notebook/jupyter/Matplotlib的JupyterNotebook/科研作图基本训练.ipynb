# 双Y轴作图

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,2.0*np.pi,101)
y=np.sin(x)
z=np.sinh(x)
#separate the figure object and axes object from the plotting object
fig,ax1= plt.subplots()

#Duplicate the axes with a different y axis and the same x axis
ax2=ax1.twinx()


ax1.plot(x,y,label="sin",color="r")
ax2.plot(x,z,label="sinh",color="b")

#Global figure properties
plt.title("plot of sin and sinh")
plt.show()
```

# Annotate Plot

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use("default")


x = np.linspace(0,2.0*np.pi,101)
C,S = np.cos(x),np.sin(x)

plt.plot(x,C,color="r",label="Cos")
plt.plot(x,S,color="g",label="Sin")

plt.plot(0.5*np.pi, np.sin(0.5*np.pi),"bo")
plt.plot(1.5*np.pi, np.sin(1.5*np.pi),"ko")
# plt.plot(0.0*np.pi, np.cos(0.0*np.pi),"ko")
# plt.plot(1.0*np.pi, np.cos(1.0*np.pi),"ko")

plt.annotate("sine local maxima",                           # text to display
            xy=(0.5*np.pi,np.sin(0.5*np.pi)),               # point to annotate
             xytext=(1.0*np.pi,np.sin(0.5*np.pi)),          # the coordinate for the text to start
             arrowprops=dict(facecolor="blue",shrink=0.1)   # arrow properties dictionary
           )

plt.annotate("sine local minma",
             xy=(1.5*np.pi,np.sin(1.5*np.pi)),
             xytext=(1.5*np.pi,np.sin(1.5*np.pi)+0.5),
             arrowprops=dict(facecolor="black",shrink=0.05)
             )

plt.grid()
plt.legend()
plt.title("Sample Sine and Cosine Curves")
plt.show()

# Adding Annotation in Plots
# https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.annotate
# For the Arrow propertites
# https:// matplotlib.org/api/patches_api.html#matplotlib.patches.FancyArrowPatch
```

# Polar Plots

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use("default")


# Parameters

n = 2001
f = np.pi
r = 5.0

# Data
theta = np.linspace(0,20.0*np.pi,n)
curve = r*np.cos(f*theta)

# Get an axes handle/obejct

ax1 = plt.subplot(111,polar=True)

# Plot the fig
ax1.plot(theta,curve,color="xkcd:salmon",label="Rose Curve")


## Main Tweaks
# Radius limits
ax1.set_ylim(-6,6)
#Radius ticks
ax1.set_yticks(np.linspace(-6,6,7))
# Radius tick position in degrees
ax1.set_rlabel_position(270)
# Angle ticks
ax1.set_xticks(np.linspace(0,2.0*np.pi,17))


# Addtional Tweaks
plt.grid(True)
plt.legend()
plt.title("Rose Plots")

plt.show()
```

# Stream  Plots

```python
import numpy as np
import matplotlib.pyplot as plt

def vander_pol_oscillator(m,x,y):
    u,v = m*(x - x**3/3 - y), x/m
    return u,v 

# half length of dimension in x and y axis
l = 8.0

# No. of points along x/y axis
n = 101
# Parameter
mu = 0.5
# X and Y parameters
x1 = np.linspace(-l,l,n)
x2 = np.linspace(-l,l,n)

#Creating a mesh field
X1,X2 = np.meshgrid(x1,x2)

# Calculating velocities
U,V = vander_pol_oscillator(mu,X1,X2)

# Resultant Velocity
vels = (U**2 + V**2)**0.5

# Line width parameter
lw=4*vels/np.max(vels)

'''
  Axes.streampot(x,y,u,v,density=1,linewidth=None,
                 color=None, cmap=None, norm=None,
                 arrowsize=1,arrowstyle="-|>",minlength=0.1,
                 transform=None,zorder=None,start_points=None)
'''


plt.figure()
stream = plt.streamplot(X1,X2,                   # x and y coordiate positions
              U,V,                      # u and v velocities
              arrowsize=2,              # arrow size
              arrowstyle = "->",        # arrow style
            #color = "xkcd:azure",   # single color,cmap not needed
              color = vels,               # color vector field, cmap needed
               density =1,               # density of streamlines
              # density=(0.5,1),        # density along x and y directions -2 value array
              linewidth =2,           # linewidth
              # linewidth=lw,
              cmap=plt.cm.jet,           # color map
              )

plt.title("Van der pol Phase Portrait")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.colorbar(stream.lines)
plt.xticks()
plt.yticks()
plt.axis([-l,l,-l,l])
plt.grid()


plt.show()

```

# Quiverplots

```python
import numpy as np
import matplotlib.pyplot as plt

def vander_pol_oscillator(m,x,y):
    u,v = m*(x - x**3/3 - y), x/m
    return u,v 

# half length of dimension in x and y axis
l = 8.0

# No. of points along x/y axis
n = 101
# Parameter
mu = 0.5
# X and Y parameters
x1 = np.linspace(-l,l,n)
x2 = np.linspace(-l,l,n)

#Creating a mesh field
X1,X2 = np.meshgrid(x1,x2)

# Calculating velocities
U,V = vander_pol_oscillator(mu,X1,X2)

# Resultant velocity
vels = np.hypot(U,V)

# Slice interval for quiver
slice_interval = 4

# Slice index for smoother quiver plot
# General note: Adjust the slice interval and scale accordingly to get
# the requried arrow size
skip = (slice(None,None,slice_interval),slice(None,None,slice_interval))

plt.figure()

Quvier = plt.quiver(X1[skip],X2[skip],
                   U[skip],V[skip],
                   vels[skip],
                   units="height",
                   angles="uv",
                   scale=250,
                   pivot="mid",
                   #color="blue",
                   cmap=plt.cm.jet, 
                   )

plt.title("Van der pol Phase Portrait")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.quiverkey(Quvier,1.01,1.01,10,label="10m/s",labelcolor="blue",labelpos="N",
             coordinates="axes")
plt.colorbar(Quvier)
plt.xticks()
plt.yticks()
plt.axis([-l,l,-l,l])
plt.grid()
plt.show()
```

# Barb Plot

```python
import numpy as np
import matplotlib.pyplot as plt

def vander_pol_oscillator(m,x,y):
    u,v = m*(x - x**3/3 - y), x/m
    return u,v 

# half length of dimension in x and y axis
l = 8.0

# No. of points along x/y axis
n = 101
# Parameter
mu = 0.5
# X and Y parameters
x1 = np.linspace(-l,l,n)
x2 = np.linspace(-l,l,n)

#Creating a mesh field
X1,X2 = np.meshgrid(x1,x2)

# Calculating velocities
U,V = vander_pol_oscillator(mu,X1,X2)

# Resultant velocity
vels = np.hypot(U,V)

# Slice interval for quiver
slice_interval = 4

# Slice index for smoother quiver plot
# General note: Adjust the slice interval and scale accordingly to get
# the requried arrow size
skip = (slice(None,None,slice_interval),slice(None,None,slice_interval))

plt.figure()
#barbs = plt.barbs(X1,X2,U,V)
barbs = plt.barbs(X1[skip],X2[skip],
                 U[skip],V[skip],
                 vels[skip],
                 pivot="tip",
                 cmap=plt.cm.jet,
                 barb_increments={"half":10,
                                  "full":20,
                                  "flag":80
                 },)

plt.colorbar(barbs)
plt.title("Van der pol Phase Portrait")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.xticks()
plt.yticks()
plt.axis([-l,l,-l,l])
plt.grid()
plt.show()

```

 # ColorMap reference

## Reference for colormaps included with Matplotlib

#### This reference example shows all colormaps included with Matplotlib.Note that
#### any colormap listed here can be reversed by appending "_r"（e.g "pink_r"）.
#### These colormaps are divided into following categories:

### Sequential

#### These colormaps are approximately monochromatic colomaps varying
#### smoothly between two color tones --- usually from low saturation(e.g white) to
#### high saturation(e.g a bright blue). Sequential colormaps are ideal for
#### representing most scientific data since they show a clear progression from
#### low- to -high values

### Diverging

#### These colormaps have a median value(usually light in color) and vary
#### smoothly to two different color tones at high and low values. 
#### Diverging colormaps are ideal when your data has a median value that is 
#### significant (e.g 0, such that positive and negative values are represented by
####  different colors of the colormap)

### Qualitative

#### These colormaps vary rapidly in color. Qualitative colormaps are useful for
#### choosing a set of discrete colors. For examples::

####  color_list =plt.cm.Set3(np.linspace(0,1,12))

#### gives a list of RGB colors that are good for plotting a series of lines on a dark 
#### dark background

### Miscellaneous

#### colormaps tha don't fit into the categories above

#### Have colormaps separated into categories:
#### http://matplotlib.org/examples/color/colormaps_reference.html
