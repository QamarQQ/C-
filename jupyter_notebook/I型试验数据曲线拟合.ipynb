```python
import pandas as pd
import numpy as np
import glob
import os
file_name=glob.glob("*-2.csv")
linear_parameter=[]
nonlinear_parameter=[]
for file in file_name:
    photo_name=os.path.splitext(file)[0]
    data=pd.read_csv(file,encoding="gbk",skiprows=3,header=None,dtype=float)
    max_index=data.index[data[3]==max(data[3])][0]
    nlinear_temp=data[max_index:]
    linear_temp=data[:max_index]
    nonlinear_fit=np.polyfit(nlinear_temp[2],nlinear_temp[3],2)
    linear_fit=np.polyfit(linear_temp[2],linear_temp[3],1)
    linear_parameter.append(tuple(linear_fit))
    nonlinear_parameter.append(tuple(nonlinear_fit))
    linear= np.poly1d(linear_fit)
    nonlinear=np.poly1d(nonlinear_fit)
    linear_yvals=linear(linear_temp[2])
    nonlinear_yvals=nonlinear(nlinear_temp[2])
    li_plot1=plt.plot(linear_temp[2], linear_temp[3], '*',label='original values')
    li_plot2=plt.plot(linear_temp[2], linear_yvals, 'r',label='polyfit values')
    no_plot1=plt.plot(nlinear_temp[2], nlinear_temp[3], '*',label='original values')
    no_plot2=plt.plot(nlinear_temp[2], nonlinear_yvals, 'r',label='polyfit values')
    plt.xlabel('位移/mm')
    plt.ylabel('载荷/N')
    plt.savefig(photo_name+"_曲线拟合"+'.png')
    plt.close()
    

    
    
```
